```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.alert import Alert
from webdriver_manager.chrome import ChromeDriverManager
import numpy as np

class Cust:
    def __init__(self, Fname, Lname, Postcode):
        self.Fname = Fname
        self.Lname = Lname
        self.Postcode = Postcode
        
Custlist = []
 
# appending instances to list
Custlist.append(Cust('Christopher', 'Connely', 'L789C349'))
Custlist.append(Cust('Frank', 'Christopher', 'A897N450'))
Custlist.append(Cust('Christopher', 'Minka', 'M098Q585'))
Custlist.append(Cust('Connely', 'Jackson', 'L789C349'))
Custlist.append(Cust('Jackson', 'Frank', 'L789C349'))
Custlist.append(Cust('Minka', 'Jackson', 'A897N450'))
Custlist.append(Cust('Jackson', 'Connely', 'L789C349'))

options = Options()
options.add_experimental_option("detach", True)

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.get("https://www.globalsqa.com/angularJs-protractor/BankingProject/#/login")
driver.maximize_window()

WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//button[normalize-space()='Bank Manager Login']"))).click()
WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//button[normalize-space()='Add Customer']"))).click()

# Accessing object value using a for loop
for obj in Custlist: 
    WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//input[@placeholder='First Name']"))).send_keys(obj.Fname)
    WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//input[@placeholder='Last Name']"))).send_keys(obj.Lname)
    WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//input[@placeholder='Post Code']"))).send_keys(obj.Postcode)
    WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//button[@type='submit']"))).click()
    alert = Alert(driver)
    alert.accept()

WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//button[normalize-space()='Customers']"))).click()

row = 6
for obj in Custlist:
    convRow = str(row)
    targetCell1 ="//tr["+convRow+"]/td[1]"
    targetCell2 ="//tr["+convRow+"]/td[2]"
    targetCell3 ="//tr["+convRow+"]/td[3]"
    checkFname = WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, targetCell1)))
    checkLname = WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, targetCell2)))
    checkPostcode = WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, targetCell3)))
    assert checkFname.text == obj.Fname
    assert checkLname.text == obj.Lname
    assert checkPostcode.text == obj.Postcode
    row = row+1

WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//tr[10]/td[5]/button"))).click()
WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//tr[6]/td[5]/button"))).click()
```

# Q2 (simplified ver.)

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.alert import Alert
from webdriver_manager.chrome import ChromeDriverManager
import numpy as np
import time

options = Options()
options.add_experimental_option("detach", True)

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.get("https://www.globalsqa.com/angularJs-protractor/BankingProject/#/login")
driver.maximize_window()


#Homepage cust login
WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//button[normalize-space()='Customer Login']"))).click()
WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//select[@id='userSelect']"))).click()
WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//option[text()='Hermoine Granger']"))).click()
WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//button[normalize-space()='Login']"))).click()
WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//option[text()='1003']"))).click()

#Deposit Function
def Deposit(depoAmount, expectedBal):
    WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//button[normalize-space()='Deposit']"))).click()
    WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//input[@placeholder='amount']"))).send_keys(depoAmount)
    WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//button[@type='submit']"))).click()
    Successful = WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//span[@class='error ng-binding']")))
    print("Response: " + Successful.text)
    assert Successful.text == "Deposit Successful"
    Balance = WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//strong[normalize-space()=" + expectedBal + "]")))
    print("Current Balance is: " + Balance.text + " rupees")
    assert Balance.text == expectedBal

#Withdrawal Function    
def Withdraw(withAmount, expectedBal):
    WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//button[normalize-space()='Withdrawl']"))).click()
    time.sleep(1)
    WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//input[@placeholder='amount']"))).send_keys(withAmount)
    WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//button[normalize-space()='Withdraw']"))).click()
    time.sleep(1)
    Successful = WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//span[@class='error ng-binding']")))
    print("Response: " + Successful.text)
    assert Successful.text == "Transaction successful"
    time.sleep(1)
    Balance = WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//strong[text()=" + expectedBal + "]")))
    print("Current Balance is: " + Balance.text + " rupees")
    assert Balance.text == expectedBal

#Calling Function
Deposit("50000", "50000")
Withdraw("3000", "47000")
Withdraw("2000", "45000")
Deposit("5000", "50000")
Withdraw("10000", "40000")
Withdraw("15000", "25000")
Deposit("1500", "26500")
    
#Logout
WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//button[normalize-space()='Logout']"))).click()
```
